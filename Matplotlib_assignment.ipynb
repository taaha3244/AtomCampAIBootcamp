```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data file
file_path = "C:/Users/taaha/Downloads/IMDB-Movie-Data.csv"
df = pd.read_csv(file_path)
```

1.	Plot a line graph showing the trend of average movie ratings over the years.

```python
avg_movie_rat=df['Rating'].mean()
years=df['Year']
```

```python
#grouping dataset by Year and taking avg
grouped_by_year=df.groupby(['Year']).mean(numeric_only=True)
grouped_by_year
```

```python
#Taking out the required columns i.e. ratings and year
avg_yr_rat=grouped_by_year['Rating']
years=df['Year'].sort_values(ascending=True).unique()
```

```python
fig,ax=plt.subplots(figsize=(6,6))
ax.plot(years,avg_yr_rat)
plt.title('Average Movie Rating Over the Years')
ax.set_xlabel('Years')
ax.set_ylabel('Avg Rating ')
plt.grid(True)
plt.show()
```

2.	Create a bar chart showing the number of movies released each year.

```python
#count of movies released by each year
df1=df.groupby(['Year'])['Title'].count()
df1.head()
```

```python
#values for the year
yr=df1.index
```

```python
#values for the count of movies
num_movie=df1.values
```

```python
fig,ax=plt.subplots(figsize=(6,6))
ax.bar(yr,num_movie)
plt.title('No Movies Released by Year')
ax.set_xlabel('Years')
ax.set_ylabel('Count of Movies ')
plt.grid(True)
plt.show()
```

3.	Scatter Plot: Create a scatter plot to analyze the relationship between 'Rating' and 'Revenue (Millions)'.

```python
rating=df['Rating']
rev=df['Revenue (Millions)']
```

```python
fig,ax=plt.subplots(figsize=(6,6))
ax.scatter(rating,rev,alpha=0.5, color='darkgreen')
plt.title('Rating and Revenue')
ax.set_xlabel('Rating')
ax.set_ylabel('Revenue (Millions) ')
plt.grid(True)
plt.show()
```

4.	Pie Chart: Make a pie chart showing the proportion of movies in different genres (consider only the primary genre).

```python
# Splitting the pandas data frame with apply and lambda function by ', ' as delimeter
df['PrimaryGenre']=df['Genre'].apply(lambda x: x.split(',')[0])
```

```python
#getting the count of movies in each category
primary_genre=df['PrimaryGenre'].value_counts()
```

```python
x=primary_genre.index
```

```python
y=primary_genre.values
```

```python
fig,ax=plt.subplots(figsize=(6,6))
ax.pie(y,labels=x)
plt.title('Movies in different genres')
plt.show()
```

6.	Stacked Bar Chart: Generate a stacked bar chart showing the number of movies in each genre over the years.

```python
genre_year_counts = df.groupby(['Year', 'PrimaryGenre']).size().unstack(fill_value=0)
```

```python
# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))
genre_year_counts.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Number of Movies in Each Genre Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

7.	Heatmap: Create a heatmap to display the correlation between numeric columns like 'Rating', 'Revenue', and 'Metascore'.

```python
corr_data=df[['Rating', 'Revenue (Millions)', 'Metascore']]
```

```python
corr_matrix=corr_data.corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Rating, Revenue, and Metascore')
plt.show()

```
